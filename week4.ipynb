# 완전탐색(Brute-Force)

- 모든 경우의 수를 탐색
- 조합(combination)과 순열(permutation) 적용 가능

## 조합, 중복조합

- 서로 다른 n개의 원소에서 r개의 원소를 뽑는 방법
- 중복 허락

## 순열

- 서로 다른 n개의 원소에서 r개의 원소를 뽑아서 나열

```python
# 중복: 중복

alpha = list('abcdefghijklmnopqrstuvwxyz')

def homogeneous(lst, r, sel=[]):
    if r==0:
        print(''.join(sel))
        return
    for data in lst:
        sel.append(data)
        homogeneous(lst, r-1, sel)
        sel.pop()
    
homogeneous(alpha, 3)
```

```python
# 순열: 뽑아서 나열

alpha = list('abcdefghijklmnopqrstuvwxyz')
한글 = list('안녕하')

def permutation(lst, r, sel=[]):
    if r==0:
        print(''.join(sel))
        return
    for i in range(len(lst)):
        temp = lst.pop(i)
        sel.append(temp)
        permutation(lst, r-1, sel)
        sel.pop()
        lst.insert(i, temp)
    
#permutation(alpha, 3)
permutation(한글, 2)
```

```python
# 조합: 뽑기만함 ##########################미완성 책으로 다시보셈

alpha = list('abcdefghijklmnopqrstuvwxyz')
한글 = list('안녕하')

def combination(lst, r, sel=[]):
    if r==0:
        print(''.join(sel))
        return
    
    for i in range(0, len(lst)):
        temp = lst.pop(i)
        sel.append(temp)
        combination(lst, r-1, sel)
        sel.pop()
        lst.insert(i, temp)
    
#combination(alpha, 3)
combination(한글, 2)
```

```python
from itertools import product
from itertools import permutations
from itertools import combinations_with_replacement
from itertools import combinations

lst = [x for x in range(1,3)]
print(len(lst))
print(lst)
print()

prod=list(product(lst,repeat=2)) # 중복순열 
print(len(prod))
print(prod)
print()

perm=list(permutations(lst,2)) # 순열
print(len(perm))
print(perm)
print()

comb_=list(combinations_with_replacement(lst,2)) # 중복조합
print(len(comb_))
print(comb_)
print()

comb=list(combinations(lst,2)) # 조합
print(len(comb))
print(comb)
print()
```

# 달팽이 리스트

https://www.acmicpc.net/problem/17827

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        
n, m, v = map(int, input().split()) # 한방에 입력받고 쪼개고 인트로 바꿈

line = input().split()

head = Node("head")
#tail = Node("tail") # 순환이므로 테일 빼버렷
prev = head

for i, data in enumerate(line): # enumerate: 열거형 인덱스와 값을 한방에 데려옴 ㄷㄷ
    node = Node(data)
    prev.next = node
    prev = node
    #node.next = tail # 순환이므로 테일 빼버렷
    if i==v-1: temp = node # 순환함 v번째 노드 잠깐 저장

node.next = temp

for _ in range(m):
    k=int(input())
    if k>=n:
        #k = ( ( k - (v-1) ) % (n-(v-1)) ) + (v-1)
        k = (k-2)%8 + 2
        if k==n:
            k=v-1

    target = head.next
    for _ in range(k):
        target = target.next
    print(target.data)
```

# 이진 트리

https://www.acmicpc.net/problem/1991



- 전위 순회한 결과 : ABDCEFG // (루트) (왼쪽 자식) (오른쪽 자식)
- 중위 순회한 결과 : DBAECFG // (왼쪽 자식) (루트) (오른쪽 자식)
- 후위 순회한 결과 : DBEGFCA // (왼쪽 자식) (오른쪽 자식) (루트)

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
        
def preorder(node): # 전위 순회
    if node == None: return
    print(node.data, end="")
    preorder(node.left)
    preorder(node.right)

def inorder(node): # 중위 순회
    if node == None: return
    inorder(node.left)
    print(node.data, end="")
    inorder(node.right)

def postorder(node): # 후위 순회
    if node == None: return
    postorder(node.left)
    postorder(node.right)
    print(node.data, end="")
    
def search(node, find_data, data_l, data_r):
    if node == None: return
    if node.data == find_data:
        if data_l != '.':
            node.left = Node(data_l)
        if data_r != '.':
            node.right = Node(data_r)
        return
    search(node.left, find_data, data_l, data_r)
    search(node.right, find_data, data_l, data_r)
    
root = Node("A")

'''node = Node("B")
root.left = node
node = Node("C")
root.right = node
node = Node("D")
root.left.left = node
node = Node("E")
root.right.left = node
node = Node("F")
root.right.right = node
node = Node("G")
root.right.right.right = node'''

n = int(input())

for _ in range(n):
    data, data_l, data_r = input().split()
    
    search(root, data, data_l, data_r)
    
preorder(root)
print()
inorder(root)
print()
postorder(root)
```

```python
import sys
sys.setrecursionlimit(100000)

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def postorder(node): # 후위 순회
    if node == None: return
    postorder(node.left)
    postorder(node.right)
    print(node.data)
    
def seat(node, data):
    if node == None: return
    if node.data > data: # 데이터가 작으면 왼쪽으로
        if node.left == None:
            node.left = Node(data)
        else:
            seat(node.left, data)
    if node.data < data:
        if node.right == None:
            node.right = Node(data)
        else:
            seat(node.right, data)
    
#input_list = sys.stdin.read()
#input_list = [50, 30, 24, 5, 28, 45, 98, 52, 60]
input_list = []

while True:
    try:
        input_list.append(int(input()))
    except:break
        
#for _ in range(9):
#    input_list.append(int(input()))

data = input_list.pop(0)
root = Node(data)

for _ in range(len(input_list)):
    data = input_list.pop(0)
    seat(root, data)

postorder(root)
```

```python
# 인접리스트 사용

n, m, v = map(int, input().split())

dic = {}

for _ in range(m):
    i, j = list(map(int, input().split()))
    if i not in dic:
        dic[i] = [j]
    else:
        dic[i] += [j]
    if j not in dic:
        dic[j] = [i]
    else:
        dic[j] += [i]


def dfs(dic, node):
    visited = [] # 방문한 노드 체크 목적
    stack = [] # 방문할 노드 체크 목적
    stack.append(node)
    
    while stack: # 스택이 빌때까지
        node = stack.pop(-1)
        if node not in visited:
            visited.append(node)
            dic[node] = sorted(dic[node])
            stack += reversed(dic[node]) # 바꿔서 스택에 넣음
            #for i in range(-1, -len(dic), -1): 
                #stack.append(dic[node][i])
    return visited


def bfs(dic, node):
    visited = []
    queue = []
    queue.append(node)
    
    while queue:
        node = queue.pop(0)
        if node not in visited:
            visited.append(node)
            dic[node] = sorted(dic[node])
            queue += dic[node]
    return visited

for ele in dfs(dic, v):
    print(ele, end=" ")
print()
for ele in bfs(dic, v):
    print(ele, end=" ")
```
